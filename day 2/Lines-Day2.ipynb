### Design Considerations
- **Theme**: A soft greige aesthetic was chosen to balance visual appeal with readability.
- **Focus on Lines**: The maps showcase road networks (motorways, primary roads, and residential streets), emphasizing the intricate web of lines that define urban spaces.
- **Neutral Palette**: The background uses a light beige-gray tone (`#e5e5e0`) to contrast against darker gray roads (`#696969`).

```python
import os
import osmnx as ox
from matplotlib import pyplot as plt
from IPython.display import Image


path = "/Users/macytrout/Documents/OSMnx"
os.makedirs(path, exist_ok=True)
os.chdir(path)


ox.config(log_file=False, log_console=True, use_cache=True)
ox.settings.cache_folder = os.path.join(path, 'cache')
os.makedirs(ox.settings.cache_folder, exist_ok=True)


img_folder = './images'
os.makedirs(img_folder, exist_ok=True)
extension = 'png'
size = 350
dpi = 900

street_widths = {'footway' : 0.5,
                 'steps' : 0.5,
                 'pedestrian' : 0.5,
                 'path' : 0.5,
                 'track' : 0.75,
                 'service' : 1.5,
                 'residential' : 2.5,
                 'primary' : 2.5,
                 'motorway' : 3}


places = {
    'London': (51.5074, -0.1278),
    'Philadelphia': (39.9526, -75.1652),
    'Paris': (48.8566, 2.3522),
    'Tokyo': (35.6895, 139.6917),
}


fig, axes = plt.subplots(1, len(places), figsize=(20, 5), constrained_layout=True)


for ax, (place, point) in zip(axes, places.items()):
    ox.plot_figure_ground(
        point=point,
        dist=1000,
        network_type='all',
        street_widths=street_widths,
        bgcolor='#e5e5e0',
        edge_color='#696969',
        figsize=(6, 6),
        ax=ax,
        show=False,
        save=False,
    )
    ax.set_title(place, fontsize=14, color='#4d4d4d', pad=10)
    ax.axis('off')


fig.patch.set_facecolor('#f5f5f0')


output_file = f"{img_folder}/combined_maps_greige.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()

print(f"Combined map saved at: {output_file}")
```
